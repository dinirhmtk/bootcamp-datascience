<h2> Jawab : </h2>

- <b>Apa itu machine learning?</b>

Machine learning adalah salah satu ilmu yang berkaitan dengan kecerdasan buatan yang memiliki tujuan untuk memberikan kemampuan belajar pada mesin. Machine learning memungkinkan sebuah mesin mempelajari banyak data untuk dapat menghasilkan proses input/output tanpa harus memakai kode program tertentu.

- <b>Apa itu data feature dan data target?</b>

Feature adalah data predictor yang dalam statistika disebut independent variable sedangkan data target atau yang dikenal dengan label dalam statistika disebut dependant variable.

- <b>Apa Perbedaan Supervised Learning dan Unsupervised Learning?</b>

Supervised learning merupakan metode belajar mesin dimana algoritma belajar menggunakan labeled data pada input maupun output. Unsupervised learning merupakan metode belajar mesin dimana algoritma belajar menggunakan no labeled data pada input maupun output. Perbedaan sederhana keduanya adalah supervised learning merupakan metode belajar berdasarkan guide dan bimbingan sedangkan unsupervised learning tidak.

- <b>Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!</b>

Jenis algoritma yang ada pada supervised learning diantaranya adalah, decision tree, naive bayes, nearest neighbor dan support vector machine. Perbedaannya diantaranya yaitu :

1. Decision tree membagi data menjadi himpunan bagian berdasarkan input
2. Naive bayes melihat data berdasarkan keterkaitan satu sama lain
3. Nearest neighbor melihat data berdasarkan kedekatan nilai atribut
4. SVM mengambil satu set data untuk memperkirakan input dan berfungsi untuk analisis data dan pengenalan pola

- <b>Apa perbedaan Hyperparameter dan Parameter?</b>

Hyperparameter merupakan konfigurasi variabel terhadap model yang didefinisikan di awal sebelum model diminta untuk mempelajari data. Sedangkan parameter merupakan variabel yang hasil estimasi dataset, nilai yang dipelajari dari data selama training. Nilai parameter model tidak diatur secara manual tetapi diperkirakan dari data training.

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# melakukan reshape dari data asli untuk data training
x_train = x.reshape(-1,1)
y_train = y.reshape(-1,1)
x_test = x_new.reshape(-1,1)
```

```python
# mendefinisikan algoritma decision tree
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor(max_depth = 6,
            min_samples_split = 2,
            min_samples_leaf = 5,
            max_leaf_nodes = 7)
```

```python
# melakukan proses training data menggunakan fitur fit
dtree.fit(x_train, y_train)

# mendefinisikan prediksi data training dan dataset
ytrain = dtree.predict(x_train)
y_pred = dtree.predict(x_test)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(x, y, color='#0000FF', label='Data Asli')
plt.scatter(x_train, ytrain,  color='#BFBF00', label='Prediksi Data Training')
plt.plot(x_test, y_pred,  color='#3F9F3F', label='Prediksi Terhadap Data Baru')

plt.legend(loc='upper right', bbox_to_anchor=(1.1, 1.2))
plt.show()
```

```python
from sklearn import tree
fig = plt.figure(figsize=(15,8))
tree.plot_tree(dtree)
```
