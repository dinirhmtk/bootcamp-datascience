```python
import numpy as np

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10,8))
x = np.linspace(-10, 10, 100)
ax.plot(x, np.cos(x), color='red', linestyle='-', label='Cosx')
ax.plot(x, np.cos(x)*3, color='green', linestyle='--', label='3 Cosx')
ax.plot(x, np.sin(x-90), color='blue', linestyle='dashdot', label='Sin(x-90)')
ax.plot(x, np.sin(x)*2, color='yellow', linestyle='none', marker='x', label='2 Sinx')

#custom axis label dan title
ax.set_xlabel('X-Axis Label')
ax.set_ylabel('Y-Axis Label')
ax.set_title('Grafik Cos Sin')
ax.legend(loc='center')

plt.show()
```
